```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://books.toscrape.com/catalogue/category/books/philosophy_7/index.html'

page = requests.get(url)

soup = BeautifulSoup(page.content, 'html')
```

```python
soup
```

```python
books = soup.find_all('article', class_='product_pod')

books
```

```python
booksScraped = soup.select('article.product_pod')

booksScraped
```

```python
print()
rating_map = {
    'One': 1,
    'Two': 2,
    'Three': 3,
    'Four': 4,
    'Five': 5
}

for book in booksScraped:
  # rating = book.p['class'][1]

  rating_string= book.find('p', class_ = 'star-rating').get('class')[1]

  rating = rating_map.get(rating_string)

  if rating >= 4:
    title = book.h3.a.text
    price = book.find('p', class_='price_color').text
    availability = book.find('p', class_='instock availability').text.strip()
    print(f"Title: {title}")
    print(f"Price: {price}")
    print(f"Rating: {rating}")
    print(f"Available?: {availability}")
    print("---")
    print()
```

```python
all_books = []

for page in range(1,51):
  url = f'https://books.toscrape.com/catalogue/page-{page}.html'
  # print(f"Scraping page {page}...")

  page = requests.get(url)

  if not page.ok:
    print(f"{page} page not found")
    continue

  soup = BeautifulSoup(page.content, 'html')

  books = soup.select('article.product_pod')

  for book in books:
    rating_string= book.find('p', class_ = 'star-rating').get('class')[1]
    rating = rating_map.get(rating_string)

    # Collect all books or only highly rated ones (optional filter)
    # if rating >= 3:

    title = book.h3.a.text
    price = book.find('p', class_='price_color').text
    availability = book.find('p', class_='instock availability').text.strip()
    all_books.append({
      'Title': title,
      'Price': price,
      'Rating': rating,
      'Availability': 'Yes' if 'In stock' in availability else 'No'
    })

df = pd.DataFrame(all_books)
print(df)
# df.to_csv('books.csv', index=False)
df.to_csv('books.csv', index=False, encoding='utf-8-sig')

print('Scraping complete, data saved to books.csv')
```
